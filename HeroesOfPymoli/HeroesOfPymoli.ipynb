```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

```python
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
total_players = len(purchase_data["SN"].value_counts())

# Displaying the result
data_players = {"Total Players" : total_players}
pd.DataFrame(data_players, index = [0])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Number of unique items
items_number = len(purchase_data["Item Name"].value_counts())

# Average price
avg_price = purchase_data["Price"].mean()

# Number of purchase
total_purchases = len(purchase_data["Item Name"])

# Total revenue
total_revenue = purchase_data["Price"].sum()

# Displaying the results in a data frame format
data_purchasing = {"Number of Unique Items" : items_number, 
                  "Average Price": f"${avg_price:.2f}",
                  "Number of Purchases": total_purchases,
                  "Total Revenue": f"${total_revenue}"}

pd.DataFrame(data_purchasing, index = [0])
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
males = purchase_data.loc[(purchase_data["Gender"]=="Male")]
females = purchase_data.loc[(purchase_data["Gender"]=="Female")]
other = purchase_data.loc[(purchase_data["Gender"]=="Other / Non-Disclosed")]

#Computing the number of players by gender
males_number = len(males["SN"].drop_duplicates())
females_number = len(females["SN"].drop_duplicates())
other_number = len(other["SN"].drop_duplicates())

# Computing the percentages
males_perc = (males_number * 100) / total_players
females_perc = (females_number * 100) / total_players
other_perc = (other_number * 100) / total_players

# Displaying the results in three data frames
data_purch_gender = {
    "Gender": ["Male", "Female", "Other / Non-Disclosed"],
    "Total Count": [males_number, females_number, other_number],
    "Percentage of Players": [f"{males_perc:.2f}%", f"{females_perc:.2f}%", f"{other_perc:.2f}%"]
}

data_gender_df = pd.DataFrame(data_purch_gender, columns = ["Gender", "Total Count", "Percentage of Players"])
data_gender_df = data_gender_df.set_index("Gender")
data_gender_df
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Average prices for each gender
fem_avg_price = females["Price"].mean()
mal_avg_price = males["Price"].mean()
oth_avg_price = other["Price"].mean()

# Total purchase value per gender
fem_tot_value = females["Price"].sum()
mal_tot_value = males["Price"].sum()
oth_tot_value = other["Price"].sum()

# Avg total purchase per person (per gender)
fem_avg_per = females["Price"].sum() / females_number
mal_avg_per = males["Price"].sum() / males_number
oth_avg_per = other["Price"].sum() / other_number

# Storing the results in a dataframe and displaying it
data_analysis_g = {
    "Gender": ["Female", "Male", "Other / Non-Disclosed"],
    "Purchase Count": [len(females), len(males), len(other)],
    "Average Purchase Price": [f"${fem_avg_price:.2f}", f"${mal_avg_price:.2f}", f"${oth_avg_price:.2f}"],
    "Total Purchase Value": [f"${fem_tot_value:.2f}", f"${mal_tot_value:.2f}", f"${oth_tot_value:.2f}"],
    "Avg Total Purchase per Person": [f"${fem_avg_per:.2f}", f"${mal_avg_per:.2f}", f"${oth_avg_per:.2f}"]
}

analysis_gender = pd.DataFrame(data_analysis_g, columns = ["Gender", "Purchase Count", "Average Purchase Price", "Total Purchase Value", "Avg Total Purchase per Person"])
analysis_gender.set_index("Gender")
```
